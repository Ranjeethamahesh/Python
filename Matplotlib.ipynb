```python
import matplotlib.pyplot as plt
import numpy as np

cities = ['Mysuru', 'Nanjangud', 'Bangluru', 'Mandya', 'Hasan']
populations = [500000, 1200000, 800000, 600000, 1500000]

plt.figure()  

plt.subplot(1, 2, 1)
plt.plot(cities, populations, color='blue', linestyle='-', marker='*')
plt.xlabel('cities')
plt.ylabel('Population')
plt.title('Population of Cities - Line Chart')

plt.subplot(1, 2, 2)
plt.bar(cities, populations, color='blue')
plt.xlabel('Cities')
plt.ylabel('Population')
plt.title('Population of Cities - Bar Chart')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

plt.figure(2, figsize=(6, 4))
plt.plot(x, y, color='blue', linestyle='--', label='Simple line')

plt.title('Simple Line', fontsize=14)
plt.xlabel('X Values', fontsize=12)
plt.ylabel('Y Values', fontsize=12)
plt.legend(fontsize=10)
plt.grid(True)

plt.show()
```

```python
print("2x2 Grid:")
plt.subplot(2, 2, 1)  
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y, color='blue')
plt.title(" Plot 1")

plt.subplot(2, 2, 2)  
plt.plot(x, np.cos(x), color='red')
plt.title("Plot 2")

plt.subplot(2, 2, 3) 
plt.scatter(x, np.random.rand(100), color='green')
plt.title("Plot 3")

plt.subplot(2, 2, 4)
plt.hist(np.random.randn(1000), bins=30, color='purple', alpha=0.7)
plt.title("Plot 4")


plt.tight_layout()

plt.show()
```

```python
import seaborn as sns
import pandas as pd
plt.figure(2, figsize=(6,4))

data = pd.DataFrame({"x": [1, 2, 3, 4, 5], "y": [10, 15, 20, 25, 30]})

sns.lineplot(data=data, x="x", y="y")
plt.title("Simple Line Plot")
plt.show()
```

```python
data = sns.load_dataset("iris")
plt.figure(2, figsize=(6,4))

sns.scatterplot(data=data, x="sepal_length", y="sepal_width", hue="species")
plt.title("Scatter Plot with Hue")
plt.show()
```
